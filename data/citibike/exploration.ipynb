```python
df["starttime"] = pd.to_datetime(df["starttime"])
```

```python
df = df[["starttime", "tripduration", "start station id", "end station id", "usertype", "birth year", "gender"]]
```

```python
# ISO: (year, week, weekday)
df["year"] = df.starttime.map(lambda x: x.isocalendar()[0])
df["week"] = df.starttime.map(lambda x: x.isocalendar()[1])
df["weekday"] = df.starttime.map(lambda x: x.isocalendar()[2])
df["hour"] = df.starttime.map(lambda x: x.hour)
df["minute"] = df.starttime.map(lambda x: x.minute)
df["duration_minutes"] = df.tripduration.map(lambda x: x // 60)
df["start_station"] = df["start station id"]
df["end_station"] = df["end station id"]
df["birth_year"] = df["birth year"].map(lambda x: x - 1920 if x > 1920 else np.NaN)
df["usertype"] = df["usertype"].map(lambda x: 0 if x == "Customer" else 1)

df = df.dropna()

df.drop(columns=["tripduration", "starttime", "start station id", "end station id"], inplace=True)
```

```python
# Slower than just mapping multiple times
def split_time(row):
    year, week, weekday = row["starttime"].isocalendar()
    row["year"] = year
    row["week"] = week
    row["weekday"] = weekday
    row["hour"] = row["starttime"].hour
    row["minute"] = row["starttime"].minute
    return row
```

```python

def year_month_iterator():
    # age/gender are present until at least Jan 2021. In 2017 the column names are different
    start_year = 2018
    end_year = 2020    
    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            name = f"{year}{month:02d}"
            yield name
```

```python
mins, maxs = [], []
for name in year_month_iterator():
    df = pd.read_csv(f"months/{name}.csv")
    mins.append(df.start_station.min())
    maxs.append(df.start_station.max())
```

```python
mins, maxs = [], []
for name in year_month_iterator():
    df = pd.read_csv(f"months/{name}.csv")
    mins.append(df.duration_minutes.min())
    maxs.append(df.duration_minutes.max())
```
